### 4. Di antara lima negara bagian teratas dalam penjualan, yaitu mode pengiriman dan kategori produk pilihan yang paling populer?

```python
df5 = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
df5.head()
```

```python
penjualan = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head().index
df6 = df[df['State'].isin(penjualan)]

pengiriman = df6.groupby('Ship Mode')['Sales'].sum()
populer_p = pengiriman.idxmax()

category = df6.groupby('Category')['Sales'].sum()
populer_c = category.idxmax()

print("Mode pengiriman terpopuler dari 5 negara bagian teratas penjualan adalah : ", populer_p)
print('Kategori produk terpopuler dari 5 negara bagian teratas penjualan adalah : ' , populer_c)
```
