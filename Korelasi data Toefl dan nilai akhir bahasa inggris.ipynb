```python
import pandas as pd
```

```python
Nilai_Toefl_IBT = [118,107,104,110,103,115,109,101,102,108,106,111,112,109,104,105,107,106,110,102]
Nilai_Akhir_Kelas = [90,74,71,77,71,85,76,70,72,76,74,78,79,77,72,73,75,73,78,70]
```

```python
df = pd.DataFrame({'x':Nilai_Toefl_IBT,'y' :Nilai_Akhir_Kelas})
data = df[['x','y']]
data
```

```python
#Mengitung nilai korelasi pearson
from scipy.stats import pearsonr
list1 = data['x']
list2 = data['y']
corr, _ = pearsonr(list1, list2)
print('Metode korelasi pearson menghasilSkan output Python yaitu : ', corr)
```

```python
#menghitung nilai korelasi spearmanr
from scipy.stats import spearmanr
list1 = data['x']
list2 = data['y']
corr, _ = spearmanr(list1, list2)
print('Metode korelasi Spearman menghasilkan output Python yaitu : ', corr)
```

```python
#menghitung nilai korelasi Tau-kendall
from scipy.stats import kendalltau
list1 = data['x']
list2 = data['y']
corr, _ = kendalltau(list1, list2)
print('Metode korelasi Tau-kendall menghasilkan output python yaitu : %.5f' % corr)
```

```python
# Mengganti nama kolom
df.rename(columns={'Nilai_Toefl_IBT': 'x', 'Nilai_Akhir_Kelas': 'y'}, inplace=True)
df.head()
```

```python
# Library untuk memunculkan Plot
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

# Untuk memunculkan Scatter Plot
plt.scatter(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot Nilai Toefl IBT vs Nilai Akhir Kelas')
plt.show()
```

```python
#Memodelkan dengan Regresi Linier Sederhana
import numpy as np
import statsmodels.api as sm
x = df[['x']]
y = df['y']
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print_model = model.summary()
print(print_model)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

# Data TOEFL dan nilai akhir kelas
toefl_scores = [118, 107, 104, 110, 103, 115, 109, 101, 102, 108, 106, 111, 112, 109, 104, 105, 107, 106, 110, 102]
english_scores = [90, 74, 71, 77, 71, 85, 76, 70, 72, 76, 74, 78, 79, 77, 72, 73, 75, 73, 78, 70]

# Visualisasi histogram
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(toefl_scores, bins=8, edgecolor='k')
plt.xlabel('Nilai TOEFL IBT')
plt.ylabel('Frekuensi')
plt.title('Histogram Nilai TOEFL')

plt.subplot(1, 2, 2)
plt.hist(english_scores, bins=8, edgecolor='k')
plt.xlabel('Nilai Akhir Kelas Bahasa Inggris')
plt.ylabel('Frekuensi')
plt.title('Histogram Nilai Akhir Kelas')
plt.tight_layout()

# Uji normalitas
shapiro_toefl = stats.shapiro(toefl_scores)
shapiro_english = stats.shapiro(english_scores)

print("Hasil uji normalitas untuk Nilai Toefl IBT adalah : ", shapiro_toefl)
print("Hasil uji normalitas untuk Nilai Akhir Bahasa Inggris adalah :",shapiro_english)

# Grafik Q-Q
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
stats.probplot(toefl_scores, dist="norm", plot=plt)
plt.title("Q-Q Plot TOEFL")

plt.subplot(1, 2, 2)
stats.probplot(english_scores, dist="norm", plot=plt)
plt.title("Q-Q Plot Nilai Akhir Bahasa Inggris")
plt.tight_layout()
plt.show()
```
